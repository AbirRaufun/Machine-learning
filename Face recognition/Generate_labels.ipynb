```python
import os
import numpy as np
from mtcnn.mtcnn import MTCNN
from keras_facenet import FaceNet
from PIL import Image

base_path = 'G:/AIUB\MAchine learning/face recog with tranfer learning'  
detector = MTCNN()
embedder = FaceNet()

embeddings = []
labels = []

def extract_face(filename):
    image = Image.open(filename).convert('RGB')
    pixels = np.asarray(image)
    results = detector.detect_faces(pixels)
    if results:
        x1, y1, w, h = results[0]['box']
        x1, y1 = abs(x1), abs(y1)
        x2, y2 = x1 + w, y1 + h
        face = pixels[y1:y2, x1:x2]
        face_image = Image.fromarray(face).resize((160,160))
        return np.asarray(face_image)
    else:
        return None

for person in ['Abir_faces', 'Shakil_faces', 'Amit_faces']:
    folder = os.path.join(base_path, person)
    for img_name in os.listdir(folder):
        img_path = os.path.join(folder, img_name)
        face = extract_face(img_path)
        if face is not None:
            emb = embedder.embeddings([face])[0]
            embeddings.append(emb)
            labels.append(person.replace('_faces', ''))

embeddings = np.asarray(embeddings)
labels = np.asarray(labels)

np.save('embeddings.npy', embeddings)
np.save('labels.npy', labels)
print(f'Generated embeddings for {len(labels)} faces')
```
